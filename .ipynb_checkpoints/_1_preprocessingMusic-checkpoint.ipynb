# Удаление из выкаченной базы watermark и сохрание в формате WAV

```python
import glob, os
import pydub
from pydub import AudioSegment
import pandas as pd
```

```python
def RemoveWatermarks(nameFolder):
    pathForClass = "/media/sf_share/pressfoto/data/" + nameFolder
    os.chdir(pathForClass)
    second = 1000
    for file in glob.glob("*.mp3"):
        sound = AudioSegment.from_mp3(pathForClass + "/" + file)
        tmp = pydub.audio_segment.AudioSegment.empty()

        index = 10000

        while((index + 17000)  < len(sound)):
            tmp = tmp + sound[index: (index + 17000)].split_to_mono()[0]

            index += 20000
        print(file)
        tmp.export("___" + file.replace("mp3", "wav"), format="wav")
```

```python
RemoveWatermarks("Sad")
```

```python
RemoveWatermarks("Tense")
```

```python
RemoveWatermarks("Angry")
```

```python
RemoveWatermarks("Happy")
```

# Разбиение на интервалы в 17 секунд и изменение имени по шаблону _[Тег класса]_[Номер трека]_[Номер интервала в треке]

```python
def SplitData(nameFolder):
    pathForClass = "/media/sf_share/pressfoto/data/" + nameFolder
    os.chdir(pathForClass)
    second = 1000
    i = 0
    for file in glob.glob("*.wav"):
        sound = AudioSegment.from_mp3(pathForClass + "/" + file)
        tmp = pydub.audio_segment.AudioSegment.empty()
        index = 10000
        
        j = 0
        while((index + 17000)  < len(sound)):
            sound[index: (index + 17000)].export("_" + nameFolder[0] + str(i) +"_" + str(j) + ".wav" , format="wav")
            index += 20000
            j += 1
        print(file)
        i += 1
```

```python
SplitData("Sad")
```

```python
SplitData("Happy")
```

```python
SplitData("Tense")
```

```python
SplitData("Angry")
```
